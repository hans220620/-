```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import time
from keras.datasets import boston_housing
from keras import models, layers as klayers
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# 載入 Boston Housing Price 數據集
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()

# 確保數據為正確形狀
print(f"原始 x_train 形狀: {x_train.shape}")  # (404, 13)
print(f"原始 x_test 形狀: {x_test.shape}")  # (102, 13)

# 進行標準化（數據歸一化）
mean = x_train.mean(axis=0)
std = x_train.std(axis=0)
x_train = (x_train - mean) / std
x_test = (x_test - mean) / std
```

```python
# 定義 MLP 模型
def build_housing_model(hidden_layers=1, neurons=64, activation='relu'):
    model = models.Sequential()
    model.add(klayers.Dense(neurons, activation=activation, input_shape=(13,)))
    for _ in range(hidden_layers - 1):
        model.add(klayers.Dense(neurons, activation=activation))
    model.add(klayers.Dense(1))  # 單輸出預測房價
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    return model
```

```python
# 測試不同架構 + 不同 batch size & epoch
configurations = [
    (1, 64, 'relu', 16, 100),
    (2, 128, 'relu', 16, 200),
    (2, 128, 'sigmoid', 32, 100),
    (2, 128, 'softplus', 32, 200)
]

results_housing = {}
training_histories = {}

plt.figure(figsize=(10, 5))
for layers, neurons, activation, batch_size, epochs in configurations:
    print(f"訓練模型: {layers} 層, {neurons} 神經元, {activation} 激活函數, batch_size={batch_size}, epochs={epochs}")
    model = build_housing_model(layers, neurons, activation)
    start_time = time.time()
    history = model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(x_test, y_test), verbose=1)
    end_time = time.time()
    test_loss, test_mae = model.evaluate(x_test, y_test, verbose=1)
    training_time = end_time - start_time
    
    y_pred = model.predict(x_test)
    test_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    test_mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    
    results_housing[f"{layers} layers, {neurons} neurons, {activation}, batch {batch_size}, {epochs} epochs"] = (test_mae, test_rmse, test_mape, training_time)
    training_histories[f"{layers}-{neurons}-{activation}"] = history
    
    plt.plot(history.history['loss'], label=f"{activation} Train Loss")
    plt.plot(history.history['val_loss'], label=f"{activation} Val Loss")

plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Boston Housing Loss Curve for Different Configurations")
plt.show()
```

```python
# 繪製誤差與訓練時間比較圖
plt.figure(figsize=(10, 5))
labels = results_housing.keys()
mae_values = [v[0] for v in results_housing.values()]
rmse_values = [v[1] for v in results_housing.values()]
mape_values = [v[2] for v in results_housing.values()]
training_times = [v[3] for v in results_housing.values()]

plt.bar(labels, mae_values, label='Test MAE')
plt.xticks(rotation=45)
plt.ylabel('Mean Absolute Error')
plt.title('Boston Housing - 測試 MAE 比較')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(labels, rmse_values, label='Test RMSE', color='green')
plt.xticks(rotation=45)
plt.ylabel('Root Mean Squared Error')
plt.title('Boston Housing - 測試 RMSE 比較')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(labels, mape_values, label='Test MAPE', color='red')
plt.xticks(rotation=45)
plt.ylabel('Mean Absolute Percentage Error')
plt.title('Boston Housing - 測試 MAPE 比較')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(labels, training_times, label='Training Time', color='orange')
plt.xticks(rotation=45)
plt.ylabel('Training Time (seconds)')
plt.title('Boston Housing - 訓練時間比較')
plt.legend()
plt.show()
```

```python
# 顯示最佳模型結果
best_model_config = min(results_housing, key=lambda x: results_housing[x][1])
best_mae, best_rmse, best_mape, best_time = results_housing[best_model_config]
print(f"最佳模型: {best_model_config}")
print(f"MAE: {best_mae:.4f}, RMSE: {best_rmse:.4f}, MAPE: {best_mape:.2f}%, 訓練時間: {best_time:.2f} 秒")
```
